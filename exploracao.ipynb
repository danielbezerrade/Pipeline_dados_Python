```python
#realizando importação da biblioteca principal
import pandas as pd
```

```python
#trazendo o arquivo

excel = "base_vendas_tecnologia_pipeline.xlsx"
dados = pd.read_excel(excel)
```

```python
#trazendo o nosso DataFrame
dados.head()
```

```python
#verificando a quantidade de linhas e colunas 

dados.shape
```

```python
#verificando o tipo de dados 

dados.dtypes
dados.info()
```

```python
#Regra para o nosso DataFrame -- retirar valores nulos do nosso dataFrame

#verificando se há dados nulos

dados.isnull().sum()
```

```python
#retirando valores nulos 
dados = dados.dropna(subset="nome_produto")
```

```python
#verificando se foi retirado os valores nulos
dados.isna().sum()
```

```python
#verificando os dados 
dados.dtypes
```

```python
dados.head()
```

```python
#realizando a alteração do tipo na coluna data
dados["data_venda"] = pd.to_datetime(dados["data_venda"],errors="coerce")
print(dados["data_venda"])
```

```python
#verificando se foi ajustado 
dados.dtypes
```

```python
#padronizando as informações das colunas
dados["id_venda"] =  dados["id_venda"].astype(str)
print(dados["id_venda"].dtype)
```

```python
dados["nome_produto"] = dados["nome_produto"].str.strip().str.capitalize()
dados["categoria"] = dados["categoria"].str.strip().str.capitalize()
dados["vendedor"] = dados["vendedor"].str.strip().str.title()
```

```python
#visualizando o nosso DataFrame
dados.head()
```

```python
dados.to_excel("dados_tratados_exploração.xlsx",index=False)
```
